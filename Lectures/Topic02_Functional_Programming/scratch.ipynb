```python
import os
import sys
import glob
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
%precision 4
```

```python
import fcm.statistics as stats
```

```python
xs1 = np.concatenate([
    np.random.multivariate_normal([0,0], np.eye(2), 100),
    np.random.multivariate_normal([5,5], np.eye(2), 900)
])
xs2 = np.concatenate([
    np.random.multivariate_normal([0,0], np.eye(2), 300),
    np.random.multivariate_normal([5,5], np.eye(2), 700)
])
xs3 = np.concatenate([
    np.random.multivariate_normal([0,0], np.eye(2), 600),
    np.random.multivariate_normal([5,5], np.eye(2), 400)
])
xs = [xs1, xs2, xs3]
m = stats.HDPMixtureModel(nclusts=8, burnin=500, niter=10, type='bem')
m.ident = True
rss = m.fit(xs, verbose=25)
rs = rss.average()
cs = rs.make_modal()

cs[0].classify(xs[0])
cs[2]
```

```python
import copy

# set reference cmap by averaging
r_consensus = copy.deepcopy(rs[0])
pis = np.sum([r.pis for r in rs], 0)
pis /= np.sum(pis)
for i in range(len(r_consensus.clusters)):
    r_consensus.clusters[i].pi = pis[i]
c_consensus = r_consensus.make_modal()
# ref_cmap = c_consensus.cmap

# for i in range(len(cs)):
#     cs[i].cmap = ref_cmap

# cs.append(c_consensus)
```
